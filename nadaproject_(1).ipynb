```python
from sklearn.svm import SVC
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer
r=LabelEncoder()
```

```python
data = pd.read_csv('train.csv')
datatest = pd.read_csv('test.csv')
data
o = data[data['stroke'] == 1]
for i in range(1,20):
    data = pd.concat([data, o], axis = 0)
```

```python
data.isna().sum()
```

```python
# data.dropna(inplace=True)
value = data['bmi'].mean()
data['bmi'].fillna(value, inplace = True)
```

```python
data.isna().sum()
```

```python
data.drop('id' , axis =1 , inplace = True)
data.drop('ever_married' , axis =1 , inplace = True)
data.drop('work_type' , axis =1 , inplace = True)
data.drop('Residence_type' , axis =1 , inplace = True)
data.gender=r.fit_transform(data.gender)
data.smoking_status=r.fit_transform(data.smoking_status)
```

```python
datatest.isna().sum()
```

```python
# datatest.dropna(inplace=True)
value = datatest['bmi'].mean()
datatest['bmi'].fillna(value, inplace = True)
```

```python
datatest.isna().sum()
```

```python
testid=datatest['id']
datatest.drop('id' , axis =1 , inplace = True)
datatest.drop('ever_married' , axis =1 , inplace = True)
datatest.drop('work_type' , axis =1 , inplace = True)
datatest.drop('Residence_type' , axis =1 , inplace = True)
datatest.gender=r.fit_transform(datatest.gender)
datatest.smoking_status=r.fit_transform(datatest.smoking_status)
```

```python
datatest
```

```python
x = data.drop('stroke' , axis = 1)
y= data['stroke'] 
```

```python
x_train , x_test , y_train , y_test = train_test_split( x , y , test_size=0.25)
clf = RandomForestClassifier(max_depth=5 ,random_state=35 )
clf.fit(x_train , y_train)
y_pred = clf.predict(x_test)
test = clf.predict(datatest)
metrics.accuracy_score(y_pred , y_test)
```

```python
pd.DataFrame({'id':testid,'stroke':test}).to_csv('predictied24.csv',index=False)
```
